```go
import (
	"sync"
	"fmt"
)

type Kernel struct {
	// 停止 kernel 执行
	stoppedChan <-chan struct{}
}

func (k *Kernel) Run() {
	var wg sync.WaitGroup
	// poll 创建 goroutine 执行独立端口监听
	poll := func(msg <-chan string, fn func(msg string)) {
		wg.Add(1)
		go func() {
			defer wg.Done()
			for {
				select {
				case k.stoppedChan:
					return
				case msg := <-ch:
					fmt.Println(msg)
				}
			}
		}()
	}
}

func main() {
	kernel := &Kernel{}
	kernel.Run()
}
```
